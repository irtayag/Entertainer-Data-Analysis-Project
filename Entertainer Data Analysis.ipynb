# Data analysis of Entertainer Breakthrough info

```python
# Year of Breakthrough/#1 Hit/Award Nomination
plt.subplot(1, 2, 1)
sns.histplot(df1['Year of Breakthrough'], bins=10, kde=True)
plt.title('Year of Breakthrough Distribution')
plt.xlabel('Year')
plt.ylabel('Count')
```

```python
# Calculate skewness and kurtosis for Birth_Year column
skewness1 = df1['Year of Breakthrough'].skew()
kurtosis1= df1['Year of Breakthrough'].kurtosis()

print("Skewness1:", skewness1)
print("Kurtosis1:", kurtosis1)
```

The distribution is approximately symmetrical with a slight left skewness.

```python
entertainers = df1['Entertainer']
breakthrough_years = df1['Year of Breakthrough']

# Plotting the bar plot
plt.figure(figsize=(12, 6))
plt.bar(entertainers, breakthrough_years, color='skyblue')
plt.xlabel('Entertainer')
plt.ylabel('Year of Breakthrough')
plt.title('Year of Breakthrough for Entertainers')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Add grid lines along the y-axis
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Assuming you have already calculated the top 10 entertainers based on some criteria
top_10_entertainers = df1['Entertainer'].value_counts().head(10).index.tolist()

# Filter the DataFrame to include only the data for the top 10 entertainers
df_top_10 = df1[df1['Entertainer'].isin(top_10_entertainers)]

# Plotting the bar plot
plt.figure(figsize=(12, 6))
plt.bar(df_top_10['Entertainer'], df_top_10['Year of Breakthrough'], color='skyblue')
plt.xlabel('Entertainer')
plt.ylabel('Year of Breakthrough')
plt.title('Year of Breakthrough for Top 10 Entertainers')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Add grid lines along the y-axis
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

The plot provides insights into the timeline of breakthrough moments within the entertainment industry for the top performers.Highest Later year of breakthrough is in year 2000 for entertainers Adele,Angelina joilie,Justin Timberlake,Keifer sutherland,Lady gaga.

```python
#Plotting the Box Plot
plt.boxplot(breakthrough_years)
plt.ylabel('Year of Breakthrough')
plt.title('Boxplot of Breakthrough Years')
plt.show()
```

The right-skewed distribution of breakthrough years indicates that most entertainers experienced success early in their careers.

```python
# Assuming you have already calculated the top 5 entertainers based on some criteria
top_5_entertainers = df1['Entertainer'].value_counts().head(5).index.tolist()

# Filter the DataFrame to include only the data for the top 5 entertainers
df_top_5 = df1[df1['Entertainer'].isin(top_5_entertainers)]

# Plotting the line plot
plt.figure(figsize=(12, 6))
for entertainer in top_5_entertainers:
    data_entertainer = df_top_5[df_top_5['Entertainer'] == entertainer]
    plt.plot(data_entertainer['Entertainer'], data_entertainer['Year of Breakthrough'], marker='o', linestyle='-', label=entertainer)

plt.xlabel('Entertainer')
plt.ylabel('Year of Breakthrough')
plt.title('Year of Breakthrough Over Time for Top 5 Entertainers')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.legend()  # Show legend
plt.grid(True)  # Add grid lines
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

Years of breakthrough over time for top 5 entertainers are 2008,1999,1962,1939,2000.

```python
#Pie chart
decade_counts = {}
for year in breakthrough_years:
    decade = str(year)[:3] + '0s'
    decade_counts[decade] = decade_counts.get(decade, 0) + 1

plt.pie(decade_counts.values(), labels=decade_counts.keys(), autopct='%1.1f%%')
plt.title('Distribution of Breakthrough Years by Decade')
plt.show()
```

The higher count of  decade is 25% in breakthrough year 1960 suggests a period of dynamic change and creative ferment within the entertainment industry, with lasting implications for cultural expression and societal influence.

# Data Analysis of Entertainer - Last work Info

```python
# Calculating average lifespan
df2['Lifespan'] = df2['Year of Death'] - df2['Year of Last Major Work (arguable)']
average_lifespan = df2['Lifespan'].mean()
print("Average Lifespan:", average_lifespan)
```

```python
#Boxplot of Lifespan
plt.figure(figsize=(8, 6))
plt.boxplot(df2['Lifespan'], vert=False)
plt.xlabel('Lifespan')
plt.title('Boxplot of Lifespan')
plt.show()
```

The right-skewed box plot of lifespan suggests that the majority of individuals have shorter lifespans, with outliers indicating a few individuals with significantly longer lifespans, highlighting variability in lifespan distribution.

```python
#Bar Plot of Year of Last Major Work
plt.figure(figsize=(10, 6))
df2['Year of Last Major Work (arguable)'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.xlabel('Year of Last Major Work')
plt.ylabel('Count')
plt.title('Bar Plot of Year of Last Major Work')
plt.show()
```

The year 1980 had the highest count of major works among all the years represented in the data.

```python
#Bar Plot for Year of Death
plt.figure(figsize=(10, 6))
df2['Year of Death'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.xlabel('Year of Death')
plt.ylabel('Count')
plt.title('Bar Plot of Year of Death')
plt.show()
```

The year 1977 had the highest count of deaths among all the years represented in the data.

```python
#Line Plot for Year of Last Major Work over Time
plt.figure(figsize=(10, 6))
plt.plot(df2['Entertainer'], df2['Year of Last Major Work (arguable)'], marker='o', color='skyblue')
plt.xlabel('Entertainer')
plt.ylabel('Year of Last Major Work')
plt.title('Line Plot: Year of Last Major Work over Time')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

 Carole Lombard's last major work occurred in a later year compared to other entertainers in the data

```python
#Scatter Plot for Year of Last Major Work vs. Year of Death
plt.figure(figsize=(10, 6))
plt.scatter(df2['Year of Last Major Work (arguable)'], df2['Year of Death'], color='skyblue', alpha=0.7)
plt.xlabel('Year of Last Major Work')
plt.ylabel('Year of Death')
plt.title('Scatter Plot: Year of Last Major Work vs. Year of Death')
plt.grid(True)
plt.show()
```

The entertainers who had their last major works later in their careers tended to live longer, showcasing a potential association between career longevity and lifespan.
